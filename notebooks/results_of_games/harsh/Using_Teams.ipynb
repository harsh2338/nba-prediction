```
df=pd.read_csv('preprocessed_scrapped_nba_games_data.csv')
df
```

```
temp=[]
for st in df['A_3PT']:
  temp.append(st[0])
s=pd.Series(temp)
s=pd.to_numeric(s)
df['A_3PT']=s.astype(float)




temp=[]
for st in df['H_3PT']:
  temp.append(st[0])
s=pd.Series(temp)
s=pd.to_numeric(s)
df['H_3PT']=s.astype(float)



temp=[]
for st in df['A_FG']:
  temp.append(st[0])
s=pd.Series(temp)
s=pd.to_numeric(s)
df['A_FG']=s.astype(float)



temp=[]
for st in df['H_FG']:
  temp.append(st[0])
s=pd.Series(temp)
s=pd.to_numeric(s)
df['H_FG']=s.astype(float)



temp=[]
for st in df['A_FT']:
  temp.append(st[0])
s=pd.Series(temp)
s=pd.to_numeric(s)
df['A_FT']=s.astype(float)



temp=[]
for st in df['H_FT']:
  temp.append(st[0])
s=pd.Series(temp)
s=pd.to_numeric(s)
df['H_FT']=s.astype(float)






```

```
df.head()
```

```
teams=df['Winner'].unique()
teams_map={}
for i in range(0,len(teams)):
  teams_map[teams[i]]=i
teams_map
df['A_Team']=df['A_Team'].map(teams_map)
df['H_Team']=df['H_Team'].map(teams_map)
df['Winner']=df['Winner'].map(teams_map)
df
```

```
train_size=int(len(df)*0.75)
```

```
dataframe=df.copy()
dataset = dataframe.sample(frac=1).values
dataset
```

```
  X = np.asarray(dataset[:,:-1])
  Y = np.asarray(dataset[:,-1])
```

```
train_size
```

```
X_train, X_test = np.asarray(X[0:train_size]), np.asarray(X[train_size:])
Y_train, Y_test = np.asarray(Y[0:train_size]), np.asarray(Y[train_size:])
```

```
lclf = LinearSVC(random_state=0, tol=1e-5, max_iter=1000)
lclf.fit(X_train, Y_train)
lclf.score(X_test,Y_test)
```

```
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, Y_train)
knn.score(X_test, Y_test)
```

```
clf = SVC(gamma='scale',probability=True)
clf.fit(X_train, Y_train)
clf.score(X_test,Y_test)
```

```
rdf = RandomForestClassifier(max_depth=8, random_state=0, n_estimators=300)
rdf.fit(X_train, Y_train)
rdf.score(X_test,Y_test)
```

```
xgbModel = XGBClassifier()
xgbModel.fit(X_train, Y_train)
xgbPredictions = xgbModel.predict(X_test)
xgbAccuracy = accuracy_score(Y_test, xgbPredictions)
```
